```julia
#stoichiometric matrix
v = zeros(Int, K, 2K)
for i = 1:K-1
    j = i
    v[i,j] = -1
    v[i,j+1] = 1
end
for i = 2:K 
    j = i+K
    v[i,j] = -1
    v[i,j-1] = 1
end
```
